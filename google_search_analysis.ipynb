```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize pytrend and define keyword
pytrend = TrendReq(hl='en=us',tz=360)
keyword = "cloud computing"

# Build payload - This was missing and caused the KeyError in 78wEr_ppKgTY
pytrend.build_payload([keyword],cat=0,timeframe='today 12-m',geo='',gprop='')

# Get and process region data - This defines 'region_data'
region_data = pytrend.interest_by_region()
region_data = region_data.sort_values(by=keyword,ascending=False).head(15)

# Original plotting code
plt.figure(figsize=(10,6))
sns.barplot(x=region_data[keyword], y=region_data.index, hue=region_data.index, palette="Blues", legend=False)
plt.title(f"Top countries searching for '{keyword}'")
plt.xlabel("Interest Value")
plt.ylabel("Countries")
plt.show()
```

```python
#world map
region_data=region_data.reset_index()
fig=px.choropleth(region_data,locations='geoName',
                  locationmode='country names',
                  color=keyword,
                  title=f"search interest for'{keyword}' by country",
                  color_continuous_scale='blues')
fig.show()
```
